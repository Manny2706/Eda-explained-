# pd.merge — quick reference

- Purpose  
  Join two DataFrame objects column-wise (like SQL JOIN).

- Basic usage
```python
import pandas as pd
merged = pd.merge(left_df, right_df, how='left', on='key_column')
```

- Key parameters
  - left, right — DataFrames to join (or use method on DataFrame: left.merge(right, ...))
  - how — one of "left", "right", "inner", "outer" (default "inner")
  - on — column name(s) present in both frames to join on
  - left_on, right_on — column(s) names if join keys differ
  - left_index, right_index — use index as join key(s) (bool)
  - suffixes — tuple like `('_x','_y')` for overlapping column names
  - indicator — if True adds column `_merge` showing source of each row
  - validate — checks merge integrity (e.g., "one_to_one", "one_to_many")

- Examples
```python
# common case: join country codes
df = pd.read_csv('zomato.csv', encoding='latin-1')
df_country = pd.read_excel('Country-Code.xlsx', engine='openpyxl')

# left join on same column name
df = pd.merge(df, df_country, on='Country Code', how='left')

# join when column names differ
df = pd.merge(df, df_country, left_on='Country Code', right_on='Code', how='left')

# keep merge provenance and check duplicates
df = pd.merge(df, df_country, on='Country Code', how='left', indicator=True)
```

- Tips / gotchas
  - If you get "KeyError" ensure the join column exists (check `df.columns` and `df_country.columns`).
  - Use `validate` to catch unexpected multiplicity errors.
  - For large joins, consider indexing join keys beforehand for performance.
