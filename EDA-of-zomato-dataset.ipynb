# pd.merge — quick reference

- Purpose  
  Join two DataFrame objects column-wise (like SQL JOIN).

- Basic usage
```python
import pandas as pd
merged = pd.merge(left_df, right_df, how='left', on='key_column')
```

- Key parameters
  - left, right — DataFrames to join (or use method on DataFrame: left.merge(right, ...))
  - how — one of "left", "right", "inner", "outer" (default "inner")
  - on — column name(s) present in both frames to join on
  - left_on, right_on — column(s) names if join keys differ
  - left_index, right_index — use index as join key(s) (bool)
  - suffixes — tuple like `('_x','_y')` for overlapping column names
  - indicator — if True adds column `_merge` showing source of each row
  - validate — checks merge integrity (e.g., "one_to_one", "one_to_many")

- Examples
```python
# common case: join country codes
df = pd.read_csv('zomato.csv', encoding='latin-1')
df_country = pd.read_excel('Country-Code.xlsx', engine='openpyxl')

# left join on same column name
df = pd.merge(df, df_country, on='Country Code', how='left')

# join when column names differ
df = pd.merge(df, df_country, left_on='Country Code', right_on='Code', how='left')

# keep merge provenance and check duplicates
df = pd.merge(df, df_country, on='Country Code', how='left', indicator=True)
```

- Tips / gotchas
  - If you get "KeyError" ensure the join column exists (check `df.columns` and `df_country.columns`).
  - Use `validate` to catch unexpected multiplicity errors.
  - For large joins, consider indexing join keys beforehand for performance.

- References
  - pandas `merge` docs: https://pandas.pydata.org/docs/reference/api/pandas.merge.html

# Pie Chart — Quick Reference

Purpose  
- Visualize proportion / percentage of categories within a whole. Best for small number of categories (<6) and when relative sizes are the main message.

When to avoid  
- Don’t use if many categories, values close in size, or precise comparison is needed — prefer a bar chart.

Key components / args (matplotlib.pyplot.pie)
- labels: category names
- sizes: numeric values (counts / sums / proportions)
- autopct: format string for showing percentages (e.g. "%1.1f%%")
- explode: offset slice(s) to highlight (tuple of floats)
- startangle: rotation of the start angle (degrees)
- shadow: add shadow (bool)
- wedgeprops / textprops: styling
- equal aspect: set ax.axis('equal') to make pie circular

Examples

- Using matplotlib.pyplot
```python
import matplotlib.pyplot as plt

labels = Country_names  # from notebook: Country_names
sizes = Country_values  # from notebook: Country_values

explode = [0.1 if i == max(sizes) else 0 for i in sizes]  # highlight largest slice

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90,
       explode=explode, shadow=True)
ax.axis('equal')  # circle
plt.title('Distribution by Country')
plt.show()
```

- Using pandas (convenient from a DataFrame / Series)
```python
# counts = df['Country'].value_counts()
counts = df['Country'].value_counts().nlargest(6)  # top 6 categories
counts.plot.pie(autopct='%1.1f%%', startangle=90, legend=False)
plt.ylabel('')  # remove ylabel
plt.title('Top 6 Countries')
plt.axis('equal')
plt.show()
```

Tips
- Combine minor categories into "Other" to reduce clutter.
- Use consistent colors and add a legend when labels overlap.
- For publication / accessibility, include exact values or a table alongside the pie.

References

# Pie Chart — quick reference

Purpose  
- Show proportions of categories within a whole. Best for few categories (≈<=6) and when relative share is the message.

When to avoid

# Explanation — group & count ratings

Code:
```python
ratings = df.groupby(
    ['Aggregate rating', 'Rating color', 'Rating text']
).size().reset_index().rename({0: 'Rating Count'})
```

What it does
- df.groupby([...]) — group rows by the three columns (one group per unique combination).  
- .size() — count rows in each group; returns a Series whose index is the grouping keys.  
- .reset_index() — convert the grouped Series into a DataFrame; grouping keys become columns and the counts become a column named `0`.  
- .rename({0: 'Rating Count'}) — rename the auto-generated count column `0` to `Rating Count`.

Resulting DataFrame columns:
- Aggregate rating
- Rating color
- Rating text
- Rating Count

Cleaner alternative (preferred):
```python
ratings = df.groupby(
    ['Aggregate rating', 'Rating color', 'Rating text']
).size().reset_index(name='Rating Count')
```

Usage:
- inspect: `ratings.head()`  
- sort: `ratings.sort_values('Rating Count', ascending=False)`  
- plot: e.g., bar/pie on `Rating Count`.

`````python
ratings=df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
`````

# sns.barplot — quick reference

Purpose  
- Create a bar plot showing an estimate (by default the mean) of a numeric variable for each level of a categorical variable. Useful for comparing summary statistics across categories.

Basic usage
```python
import seaborn as sns
sns.barplot(x='category_col', y='numeric_col', data=df)
```

Key parameters
- data: DataFrame
- x, y: column names (x usually categorical, y numeric). If y is an already-aggregated statistic (e.g., counts), barplot will plot those values.
- hue: additional categorical variable to produce grouped bars
- estimator: aggregation function (default = np.mean). Example: estimator=np.sum
- ci: size of confidence interval (default 95). Set ci=None to hide error bars.
- order / hue_order: specify categorical ordering
- palette: color palette (list or named palette)
- dodge: bool, whether to separate hue levels (default True)
- capsize: width of error bar caps (float)
- orient: "v" or "h" to force orientation
- ax: matplotlib Axes to draw on

Examples
```python
# simple mean-by-category
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, ci=None, palette='viridis')

# grouped bars by hue with custom palette and capsize
sns.barplot(
    x='Aggregate rating', y='Rating Count', hue='Rating color',
    data=ratings, palette=['grey','red','orange','yellow','green','darkgreen'],
    ci=None, capsize=0.05, dodge=True
)
```

Notes & tips
- By default seaborn computes an aggregate (mean) and shows confidence intervals; if your data is already aggregated (counts/sums) set ci=None.
- If categories are many or ordered, use order=... to control x-axis order.
- For large datasets consider using estimator=np.median or plot raw points with sns.stripplot / sns.swarmplot underneath.
- Use plt.figure(figsize=(w,h)) before calling sns.barplot to control plot size.
- For stacked bars use DataFrame pivot + matplotlib directly (seaborn does grouped, not stacked, bars).

References  
- seaborn.barplot docs: https://seaborn.pydata.org/generated/seaborn.barplot.html
